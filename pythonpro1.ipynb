```python
#dropping blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

EXPLORATORY DATA ANALYSIS

GENDER

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax = sns.countplot(x='Gender',data=df)      #gives count of no. of male and female
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False
```

```python
salesbygender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=salesbygender)
```

From above graphs we can say that most of the buyers are women and they also spent more.

AGE

```python
df.columns
```

```python
df.head()
```

```python
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
salesbyage = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=salesbyage)
```

From above graph we can say that most of the buyers are in the age group 26-35 and are female.

STATE

```python
df.columns
```

```python
#
sale_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='State',y='Orders',data=sale_state)
sns.set(rc={'figure.figsize':(15,5)})
```

From above graphs we can see that most of the orders and totalsales are from UP,Mharashtra and Karnataka respectively.

MARITAL STATUS

```python
df.columns
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data = sales,x='Marital_Status',y='Amount',hue='Gender')
sns.set(rc={'figure.figsize':(5,5)})
```

From above graphs most of the buyers are married(women) and have high purchasing power.

OCCUPATION

```python
ax = sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(200,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sale=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sale,x='Occupation',y='Amount')
sns.set(rc={'figure.figsize':(200,5)})
```

From above graph we can see that most of the buyers are working in IT,Heathcare and Aviation sectors

PRODUCT CATEGORY

```python
ax  = sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(200,50)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales,x='Product_Category',y='Amount')
sns.set(rc={'figure.figsize':(200,50)})
```

From above graphs we can see that most of the sold products are from Food,Clothing and Electronics category.

```python
sales=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales,x='Product_ID',y='Amount')
sns.set(rc={'figure.figsize':(200,50)})
```

CONCLUSION

"Married women in age group 26-35 yrs from UP,Maharashtra and Karnataka working in IT,Heathcare and Aviationn are more likely to buy products from Food,Clothing and Electronics category".
